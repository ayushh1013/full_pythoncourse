```python
class BMW:
    def __init__(self,gaddi_color, your_name,model): #initialization
         self.wheel = 4 #attributes
         self.spare_wheel = 1
         self.color = gaddi_color
         print('welcome ',your_name)
         print('your bmw  is ready to drive') 
         print('you bmw model is ',model)

    # Methods
    def engine_start(self):
        print('engine start vroom vroom ðŸ¤«!!')
    def turbo_charger(self):
        print('performance increase')

    def first_gear(self):
        print('car in first gear')
        
    def stop_engine(self):
        print('engine off')
    def apply_brake(self):
        print('brakes applied')

    def horn(self):
        print('pee pee')
    def auto_lock(self):
        print('car lock')
        
```
