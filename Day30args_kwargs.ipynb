```python
# global variable
```

```python
country = 'india'

def about():
    global country  
    country = 'nepal'
    print(f' I love my {country}')

print('first line: ',country)

about()
print('second line: ',country)
```

```python
def one():
    country = 'india'
    print(f'i love my {country} by one func')
    def two():
        country = 'nepal'
        print(f' i love my {country}')

    print('first line: ', country)
    two()
```

```python
one()
```

```python
# nonlocal variable
```

```python
def one():
    country = 'india'
    print(f'i love my {country} by one func')
    def two():
        nonlocal country
        country = 'nepal'
        print(f' i love my {country}')
    two()
    print('first line: ', country)
    
```

```python
one()
```

```python
def check_prime(num):
    for i in range(2,num):
        if num%i == 0:
            break

    else:
        return num
```

```python
check_prime(10)
```

```python
def give_prime(number):
    result = []
    for i in range(2,number+1):
        if check_prime(i) != None:
            result.append(check_prime(i))

    return result
```

```python
print(give_prime(1000))
```
