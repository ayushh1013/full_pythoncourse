```python
def summ(data:'list,tuple'):
    result = 0
    for i in data:
        result +=i

    return result
```

```python
def factorial(num:int):
    result = 1
    for i in range (1,num+1):
        result *=i

    return result
```

```python
factorial(9)
```

```python
def count_char(string,char):
    count = 0
    for i in string:
        if i == char:
            count += 1
    
    return count
```

```python
count_char('japan','a')
```

```python
def most_rept_char(string_data):
  

    
    dict_data = {}
    for i in string_data:
        dict_data[i] = count_char(string_data,i)
    
    my_max = 0
    most_char = ''
    for key,value in dict_data.items():
       if value>my_max:
           my_max = value
           most_char = key
    
    return {most_char:my_max}
```

```python
def most_rept_char(string_data):
  

    
    dict_data = {}
    for i in string_data:
        dict_data[i] = count_char(string_data,i)
    my_max = 0 
    most_char_index = 0
    for index,j  in enumerate(dict_data.values()):
        if j > my_max:
            my_max = j
            most_char_index = index
    return my_max,list(dict_data)[most_char_index]
```

```python
import string
import random
def generate_password(length:int):
    data =  string.ascii_uppercase + string.ascii_lowercase+string.digits
    pwd = ''
    for i in range(length):
    
        pwd += random.choice(data)

    return pwd
```

```python
generate_password(10)
```
