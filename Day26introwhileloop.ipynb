```python
# iterate dictionary using while loop

data = {'name':'karen','age':23,'class':12,'school':'pvt school','address':'noda'}
keys = data.keys()
index = 0
while index<len(data):

    print(f'your {list(keys)[index]} is: {data[list(keys)[index]]}')
    index +=1
```

```python
# common element
l_1 = [1,24,321,23,31,34,232]
l_2 = [23,4324,54,54,52,324,]

common = []

for i in l_1:
    for j in l_2:
        if i ==j:
            common.append(i)

print(common)
```
