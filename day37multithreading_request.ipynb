```python
import threading as th
```

```python
import requests as r
```

```python
start = time.time()
for i in range(1,10):
  url = '''https://picsum.photos/200/300'''
  response = r.get(url)
  data = response.content

  with open(f"C:\\Users\\ASUS\\OneDrive\\Desktop\\ayush.jpg\\image_{i}.jpeg",'wb') as f:
     f.write(data)

end = time.time()
ttt = end - start
print('total time taken: ',ttt)
```

```python
def give_image(name):
    url = '''https://picsum.photos/200/300'''
    response = r.get(url)
    data = response.content

    with open(f"C:\\Users\\ASUS\\OneDrive\\Desktop\\ayush.jpg\\image_{name}.jpeg",'wb') as f:
       f.write(data)
```

```python
give_image('ayush')
```

```python
t1 = th.Thread(target=give_image,args=['ayush'])
```

```python
t1.start()
```

```python
all_thread = []

for i in range(1,20):
    t = th.Thread(target=give_image,args=[i])
    all_thread.append(t)

for j in all_thread:
    j.start()
```
