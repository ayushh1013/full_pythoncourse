# Encapsulation

```python
#public variable
class ATM:
    def __init__(self):
        self.pin = '0000'
        self.balance = 323543543
    def deposit(self):
        print('deposit')

    def play_audio(self):
        print('audio')
    def withdrawl(self):
        print('withdrawl')
        
```

```python
axis = ATM()
```

```python
axis.withdrawl
```

```python
axis.balance
```

```python
#protected variable
class ATM:
    def __init__(self):
        self.__pin = '0000'
        self.__balance = 323543543
    def deposit(self):
        print('deposit')

    def play_audio(self):
        print('audio')
    def withdrawl(self):
        print('withdrawl')
        
```

```python
axis = ATM()
```

```python
axis._balance
```

```python
ATM()._ATM__pin = 9098
```

```python
fbi = ATM()
```

```python
fbi._ATM__pin
```
