```python
data = 'we have different phone number like 8817398398,948279982,hfd74,9874298749,892734984,some post code'

phone_numbers = []

temp_num = ''
for i in data:
    if i.isdigit():
        temp_num+= i

    if len(temp_num)== 10:
        phone_numbers.append(int(temp_num))
        temp_num = ''

print(phone_numbers)
```

## regular expression

pattern  matching

```python
import re
```

```python
data = 'we have different phone number like 8817398398,948279982,hfd74,9874298749,+91-892734984,some post code'
pt = r'\+?9?[1]?-?\d{3,12}'

re.findall(pt,data)
```

```python
#meta characters
#*+-?.()[]{}!@#$%^_/d,/D,/w/W,9,0
```

```python
data = 'we are un noida sec 15'
pt = r'\d' # it represent digit
result = re.findall(pt,data)
```
