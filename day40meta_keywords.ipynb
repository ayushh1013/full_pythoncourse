```python
import re

data = ' i love arry zarrrry jarrry marrrry karrry parry 1arrrrry 10arrrry'

pt = r'[^a-zA-Z]ar+y'
result = re.findall(pt,data)  #^ means not peak alphabets
print(result)
```

```python
data = '''Note1: we love python ds'''

pt = r'Note1: \w{1,}.+'

result = re.findall(pt,data)
print(result)
```

```python
data = '''Note1: we love python ds'''

pt = r'Note1: (?:\w{1,}.+)'

result = re.findall(pt,data)
print(result)
```
