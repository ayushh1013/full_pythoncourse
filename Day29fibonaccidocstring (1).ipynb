```python
def minn(data:'list,tuple'): #docstring
    '''this is a maxx docstring 
    ex: maxx([1,223,3,43,4324,4,23])
    ans: 432'''
    my_minn = data[0]
    for i in data:
        if i<my_minn:
            my_minn = i

    return my_minn
    
```

# fibonacci

```python
fibbo = [0,1]
no  = 5
for i in range(no-2):
    fibbo.append(fibbo[-1] + fibbo[-2])

print(fibbo)
```

```python
def get_fibbo(howmuch:'int'):
    '''To get numbers based on given length'''
    if type(howmuch) == int:
      fibbo = [0,1]
      for i in range(howmuch-2):
        fibbo.append(fibbo[-1] + fibbo[-2])

      return fibbo

    else:
        return None  
```

```python
get_fibbo(10)
```

```python
print(get_fibbo(10)+[123,3212])
```

```python
get_fibbo(28.1)
```

## checking anagram

```python
# write a funcyion which takes two words and check they aare anagrams or not
```

```python
def check_anagram(str1:'str',str2:'str') -> str:
    '''to check the anagram'''
    data1 = {}
    data2 = {}
    for i in str1:
        data1[i] = str1.count(i)

    for j in str2:
        data2[j] = str2.count(j)

    return 'anagram' if  data1 == data2 else "not an anagram"
```

```python
check_anagram('ayush','drishti')
```

```python

check_anagram('silent','listen')
```

```python
check_anagram('angel','angel')
```

## check palindrome

```python
def check_palindrome(string:'str'):
    return 'palindrome' if string.lower() == string.lower().lower()[::-1] else 'not palindrome'
```

```python
check_palindrome('oyo')
```

```python
check_palindrome('ayush')
```

## check armstrong

```python
def check_armstrong(number:int):
    '''to check the armstrong'''
    lt = len(str(number))
    result = 0
    for i in str(number):
        result += int(i)**lt

    return 'Armstrong' if result == number else 'not an armstrong'
        
```

```python
check_armstrong(9800817)
```

```python
check_armstrong(2124)
```

```python
def is_prime(n):
    return n > 1 and all(n % i for i in range(2, int(n**0.5) + 1))
```
