```python
def make_connection():
    try:
        conn = py.connect(user = 'root', host='localhost',password='1401',autocommit = True)
        cur = conn.cursor()
        return cur
    except:
        return False
```

```python
cur = make_connection()
```

```python
def execute_query(query):
    try:
        cur.execute(query)
        result = cur.fetchall()
        if len(result)>0:
            all_columns = []
            for i in cur.description:
                all_columns.append(i[0])
            df = pd.DataFrame(result,columns = all_columns)
            return df
        else:
            return'query executed '
    except Exception as e:
        print(e)


```

```python
execute_query('select * from payments.payments;')
```

```python
conn.close()
```

```python
cur = make_connection()
if not cur:
    print('connection failed!')

else:
    print('connection established!')
```
