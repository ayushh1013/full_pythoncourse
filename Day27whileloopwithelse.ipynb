```python
#while with else

a = [12,233,2335,34,4,322]
c = 0 
while c<len(a):
    if a[c] == 34:
        c+=1
        continue


    else:
        print(a[c],a[c]**2)
        c+=1
        


```

```python
points = 0
import random
round_details = {}
r = 1
while True:
    user = input('guess a number: ')
    comp = str(random.randint(1,10))
    print('comp choice:',comp)
    round_details[f'Round_{r}'] = {'comp':comp,'user':user}
    if user == comp:
        print('you are wright!!',end = '\n'*2)
        points += 10

    elif user.lower() in ['exit','e','q']:
        print('Final winning points: ' ,points)
        break

    else:
        print('you are wrong!!\n try again',end = '\n'*2)
        if points>=5:
            points -= 5
    display(clear=True)
```
