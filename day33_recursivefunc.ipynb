```python
def factorial(num):
    if num == 1:
        return num

    else:
        return factorial(num-1)*num
```

```python
factorial(3)
```

```python
# factorial(3)*4
# factorial(3-1)*3*4
# factorial(2-1)*2*3*4
# #1*2*3*4
```

```python
def sumofn(num):
    if num == 1:
        return num

    else:
        return sumofn(num-1)+num
```

```python
sumofn(10)
```

```python
#decorator

def say_hello(func):
    print('i am ayush')
    func()

    print('we are in jabalpur')
    
```

```python
#yield
def give_one_byone():
    yield 'ayush'
    yield 'uwfuho'
    yield [13,321,213,213,23]
    yield True
```

```python
obj = give_one_byone()
```

```python
obj
```

```python
next(obj)
```

```python
def a():
    print('i am a')

def b():
    print('i am  b')


def generate():
    yield ([1,12,2,23,2])
    yield a()
    yield b()
    yield('i am great')
```

```python
ob = generate()
```

```python
ob
```

```python
next(ob)
```
