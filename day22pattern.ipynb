### fibonacci

```python
fibo = [0,1]
user = int(input())

for i in range(user-len(fibo)):
    fibo.append(fibo[-1]+fibo[-2])

print(fibo)
```

## Pattern Printing

```python
no = 5
pt = '* '

for i in range(1,no+1):
    print(pt*i)
```

```python
no = 5
pt = '* '

for i in range(no,0,-1):
    print(pt*i)
```

```python
data = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
no = 4
for i in range(no):
    print(data[0:i+1])
```

```python
data = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
no = 4
for i in range(no):
    print(' '.join(data[0:i+1]))
```
