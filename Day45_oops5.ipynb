```python
class Animal:
    def speak(self):
        print('many different speaks nature!!')
class Dog:
    def speak(self):
        print('Bark !! Bark !!')
class Cat(Animal):
    def Speak(self):
        print('Meow!! Meow!!')
```

```python
ob1 = Animal()
ob1.speak()
```

```python
ob2 = Cat()
```

```python
ob2.Speak() #always prefernce thier own function
```

```python
from abc import ABC,abstractmethod
```

```python
class kejri(ABC):
    @abstractmethod
    def lake_city(self):
        print('city of lake!!')
    def pm_candidate(self):
        print('Dream to become a PM')
    def yamuna_clean(self):
        print('want to clean yamuna river!!')
```

```python
delhi = kejri()
print('object created successfully')
```
