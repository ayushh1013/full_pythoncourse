```python
class My_set:
    def __init__(self,data):
        self.data = data
    def __add__(self,other):
        return set(list(self.data) + list(other.data))
```

```python
my_a = My_set({1,24})
my_b = My_set({1,243,5})

my_a + my_b
```

# Getter

```python
# getter method :- data show to read only
# Public
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        print(f''' your name is: {self.name}\n your salary is {self.salary}''')
```

```python
e1= Employee('ayushh',68763)
e1.get_details()
```

```python
# getter method :- data show to read only
# Protected
class Employee:
    def __init__(self,name,salary):
        self._name = name
        self._salary = salary

    def get_details(self):
        print(f''' your name is: {self._name}\n your salary is {self._salary}''')
```

```python
e1= Employee('ayush',68763)
e1.get_details()
```
